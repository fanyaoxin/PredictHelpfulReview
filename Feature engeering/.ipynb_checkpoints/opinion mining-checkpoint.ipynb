```python
def compute_aspect_value(business_ids):
    global df_review
    global aspect_l
    d={}
    df_inter=df_review[df_review.business_id.isin(business_ids)]
    for name,group in tqdm(df_inter.groupby('business_id')):
        reviews=group.text.values
        aspect_value={}
        for j in range(len(aspect_l)):
            l=dict(([(aspect, score_aspect(reviews, aspect)) for aspect in aspect_l[j]]))
            l=[l[i] for i in l]
            value=np.nansum(l)/len(l)
            print(value)
            aspect_value[j]=value
        d[name]=aspect_value
    return d
    
    
```
