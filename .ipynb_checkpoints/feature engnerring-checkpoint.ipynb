```python
##construct other text features
from nltk.corpus import stopwords
import nltk
import time
from collections import Counter
from nltk import pos_tag
from nltk import word_tokenize

lem=nltk.wordnet.WordNetLemmatizer()
def text_processing(df):
    start=time.clock()
    df['text']=[' '.join(i.split()) for i in df.text]
    print('1/4', float(time.clock()-start))
    df['text']=[i.lower() for i in df.text]
    print('2/4', float(time.clock()-start))
    df['filtered_text']=[' '.join([i for i in j.split() if i not in stopwords.words('english') ]) for j in df.text]
    print('3/4', float(time.clock()-start))
    df['lemmatize_text']=[' '.join([lem.lemmatize(i) for i in j.split() ]) for j in df.text]
    print('4/4', float(time.clock()-start))
    df['lemmatize_filtered_text']=[' '.join([lem.lemmatize(i) for i in j.split() ]) for j in df.filtered_text]
    print('4/4', float(time.clock()-start))
    
    

def construct_other_text_features(df):
    
    df['review_all_length']=[len(i.split(' ')) for i in df.text]
    df['review_exceptstop_length']=[len(i.split(' ')) for i in df.filtered_text]


##constructing text
def compute_tag(text):
    l=Counter([j for i,j in pos_tag(word_tokenize(text))])
    num=sum([l[i] for i in ['NN','NNP','NNS','NNPS']])
    adj=sum([l[i] for i in ['JJ','JJR','JRS']])
    adv=sum([l[i] for i in ['RB','RBR','RBS']])
    verb=sum([l[i] for i in ['VB','VBD','VBG','VBN','VBP','VBZ']])
    return list([num,adj,adv,verb])

    
def construct_words_tag_feature(df):
    df['noun']=[compute_tag(i)[0] for i in df['lemmatize_filtered_text']]
    df['adj']=[compute_tag(i)[1] for i in df['lemmatize_filtered_text']]
    df['adv']=[compute_tag(i)[2] for i in df['lemmatize_filtered_text']]
    df['verb']=[compute_tag(i)[3] for i in df['lemmatize_filtered_text']]
    
```

```python
text_processing(df)
print('processing finish')
df.ix[:2]
```

```python
print('construct length')
construct_other_text_features(df)
df.ix[:1]
```

```python
print('construct words_tag')
construct_words_tag_feature(df)
df.ix[:2]
```

```python
import curses 
from curses.ascii import isdigit 
import nltk
from nltk.corpus import cmudict 
d = cmudict.dict() 
def nsyl(word): 
    return [len(list(y for y in x if isdigit(y[-1]))) for x in d[word.lower()]][0]
```

```python
from textstat.textstat import textstat
import re

def compute_syllables(text):
    num_sentence=textstat.sentence_count(text)
    text=re.sub('[^A-Za-z0-9]+', ' ', text)
    word_list=text.split()
    num_simple=0
    num_complex=0
    num_syllables=0
    for i in word_list:
        try:
            syllables=nsyl(i)
            if syllables>=3:
                num_complex+=1
                num_syllables=num_syllables+syllables
            else:
                num_simple+=1
                num_syllables=num_syllables+syllables
        except:
            continue
    return [num_simple,num_complex,num_syllables,num_sentence]

        
```
